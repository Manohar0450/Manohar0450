```python
from types import MethodDescriptorType
import pandas as vk
import numpy as ks
dataset = vk.read_csv("/content/age_salary_dataset.csv")
dataset
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
print("independent variables",x)
print(y)
```

```python

from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=ks.nan,strategy="mean")
x=imp.fit_transform(x)
y=y.reshape(-1,1)
y=imp.fit_transform(y)
y=y.reshape(-1)
print(x)
print(y)
```

```python
x_df = vk.DataFrame(x, columns=dataset.columns[:-1])
y_df = vk.DataFrame(y, columns=[dataset.columns[-1]])
combined_data = vk.concat([x_df, y_df], axis=1)
print("Combined data is:", combined_data)
```

```python

combined_df=vk.DataFrame(combined_data,columns=list(dataset.columns))
combined_df.to_excel("/content/combined_data.xlsx",index=False)
print("combined Dataframe",combined_df.head())
```
